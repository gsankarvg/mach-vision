```python
import cv2
import matplotlib.pyplot as plt
```

1. Import the opencv model in python load images, display images, save and get the load image dimensions.

```python
image = cv2.imread('Road.jpg')  # Provide your image file path
dimensions = image.shape
print(f"Dimensions are Height, Width, Channels = {dimensions}")
# Convert the image from BGR (OpenCV default) to RGB (Matplotlib default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.show()
```

2. Convert a colour image to grayscale. Display the change in dimensions between grayscale and colour images.

```python
image = cv2.imread('Flowers.jpg')
org_dimensions = image.shape
print(f"Original dimensions are Height, Width, Channels = {org_dimensions}")
grayscale_img = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
grayscale_img_dim = grayscale_img.shape
print(f"Gray Scale image dimensions are Height, Width, Channels ={grayscale_img_dim}")
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title("Color Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(grayscale_img, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')

plt.show()
```

3. View the indivitual channels of RGB images.

```python
image = cv2.imread('Valley.jpg')
(blue, green, red) = cv2.split(image)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20, 13))
    
# Red channel
plt.subplot(1, 3, 1)
plt.imshow(red, cmap='Reds')
plt.title("Red Channel")
plt.axis('off')

# Green channel
plt.subplot(1, 3, 2)
plt.imshow(green, cmap='Greens')
plt.title("Green Channel")
plt.axis('off')

# Blue channel
plt.subplot(1, 3, 3)
plt.imshow(blue, cmap='Blues')
plt.title("Blue Channel")
plt.axis('off')

# Show the individual channel images
plt.show()
```
